```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("../data/03-lang.csv")
```

```python
tqdm.pandas()
```

```python
count = len(df.index)
print(count)
```

```python
df = df.dropna(subset=["Text"])
print(len(df.index))
```

```python
df = df.drop(df[df.lang != 'sv'].index)
print(len(df.index))
```

```python
df['wordcount'] = df.progress_apply(lambda row: len(row['Text'].split()), axis=1)
```

```python
df = df.drop(df[df.wordcount <= 5].index)
print(len(df.index))
```

```python
df.to_csv('../data/04-filtered.csv', index=False)
```
